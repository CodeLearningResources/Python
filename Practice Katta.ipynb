```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x*2+23
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Demo')
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x,y,'r')
plt.title('First')
plt.subplot(2,3,2)
plt.plot(x,y,'b')
plt.title('Second')
plt.subplot(2,3,3)
plt.plot(x,y-3,'g')
plt.title('Third')
```

```python
fig,axes=plt.subplots(nrows=1, ncols=3)
axes[0].plot(x,y)
axes[0].set_title(' frist Plot')
axes[1].plot(x,y)
axes[1].set_title(' Scond Plot')
axes[2].plot(x,y)
plt.tight_layout()
```
