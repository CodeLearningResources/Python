```python
#Box plot
import seaborn as sns
```

```python
sns.set(style='whitegrid')
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

#steps to plot the box plot
'''
1. sort the in ascending order
2. find the median of the sorted data 
3. again find the median of left side and right side
4. we divide the data in to 4 quartiel here from start to 1 st median= 1 quartile
5. 1st median to 2nd median=2 quartile 
6. 2nd median to 3rd median=3 quartile
7. 3rd quartile to last data point= 4th quartile
8. take 2nd and 3rd quartile and plot the box remaing will be the tails of the plot

''' 

```python
sns.boxplot(tips['total_bill'])
print(tips['total_bill'].median())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,8))
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Day and Total bill")
#this grouped by the day categorical value
```

```python
plt.figure(figsize=(14,8))
#nested grouping by two categorical variable
sns.boxplot(x='day', y='total_bill',hue='smoker',data=tips)
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x='day',y='total_bill',hue='time',data=tips,linewidth=2.5)
```

```python
sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```
